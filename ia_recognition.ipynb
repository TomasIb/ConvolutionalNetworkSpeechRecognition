```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os

file_paths = []
labels = []
words_spoken = []

labels_test = []
file_paths_test = []
data_subset=['no', 'yes']
```

```python

for f in os.listdir('train'):
    
    if f in data_subset:

        for w in os.listdir('train/' + f):
            file_paths.append('train/' + f + '/' + w)
            labels.append(f)
            if f not in words_spoken:
                words_spoken.append(f)



for f in os.listdir('test'):
    if f in data_subset:
        for w in os.listdir('test/' + f):
            file_paths_test.append('test/'+f+'/'+w)
            labels_test.append(f)
       

```

```python
#train
from scipy.io import wavfile

data_train = np.zeros((len(file_paths), 32000))
maxsize = -1
for n,file in enumerate(file_paths):
    _, d = wavfile.read(file)
    data_train[n, :d.shape[0]] = d
    if d.shape[0] > maxsize:
        maxsize = d.shape[0]
    
data_train = data_train[:, :maxsize]

    
#Each sample file is one row in data, and has one entry in labels
print('Number of files total:', data_train.shape)
all_labels = np.zeros(data_train.shape[0])
for n, l in enumerate(set(labels)):
    all_labels[np.array([i for i, _ in enumerate(labels) if _ == l])] = n

print('Labels and label indices', all_labels)
print(labels)





```

```python
#TEST


data_test = np.zeros((len(file_paths_test), 32000))
maxsize = -1
for n,file in enumerate(file_paths_test):
    _, d = wavfile.read(file)
    data_test[n, :d.shape[0]] = d
    if d.shape[0] > maxsize:
        maxsize = d.shape[0]
    
data_test = data_test[:, :maxsize]

    
#Each sample file is one row in data, and has one entry in labels
print('Number of files total:', data_test.shape)
all_labels_test = np.zeros(data_test.shape[0])
for n, l in enumerate(set(labels_test)):
    all_labels_test[np.array([i for i, _ in enumerate(labels_test) if _ == l])] = n

print('Labels and label indices', all_labels_test)


```

```python
import scipy

def stft(x, fftsize=64, overlap_pct=.5):   
    #Modified from http://stackoverflow.com/questions/2459295/stft-and-istft-in-python
    hop = int(fftsize * (1 - overlap_pct))
    w = scipy.hanning(fftsize + 1)[:-1]    
    raw = np.array([np.fft.rfft(w * x[i:i + fftsize]) for i in range(0, len(x) - fftsize, hop)])
    return raw[:, :(fftsize // 2)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

spectrograms = []

for i in range(data_train.shape[0]):
  
    log_freq = 20 * np.log(np.abs(stft(data_train[i, :])) + 1)
       
    spectrograms.append(log_freq)
    print("Calculando fft : ",i)
spectrograms = np.atleast_3d(spectrograms)

```

```python
spectrograms_test = []

for i in range(data_test.shape[0]):
  
    log_freq = 20 * np.log(np.abs(stft(data_test[i, :])) + 1)
       
    spectrograms_test.append(log_freq)
    print("Calculando fft : ",i)
spectrograms_test = np.atleast_3d(spectrograms_test)
```

```python
speaker= [100, 2200,3600,3935]
i=1
for s in speaker:
    plt.figure(1, figsize=(30, 30))
    plt.subplot(2, 4, i)
    plt.imshow(spectrograms[s], cmap='gray', interpolation=None) 
    plt.xlabel('Freq')
    plt.ylabel('Time')
    #plt.ylim([1])
    plt.title('PSD  %s '%labels[s]) 
    i=i+1
```

```python
X_train = spectrograms.astype('float32')
y_train = all_labels

X_test = spectrograms_test.astype('float32')
y_test = all_labels_test



```

```python
#Normalizacion
X_train , X_test= X_train / 255.0 , X_test / 255.0
```

```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from tensorflow.keras import datasets, layers,models
from tensorflow.keras.layers import Dense, Conv2D, Flatten

def conv_net():
  model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, 4,4, input_shape = (498,32,1), activation = 'relu'),  
    tf.keras.layers.MaxPooling2D(pool_size = (2, 2)),
    tf.keras.layers.Conv2D(64, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size = (2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation = 'sigmoid'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
  ])
  return model

model=conv_net()
model.compile(optimizer='adam',
                loss='binary_crossentropy',
                metrics=['accuracy'])
```

```python
X_train = X_train.reshape(( X_train.shape[0], 498, 32,1))
X_test = X_test.reshape(( X_test.shape[0], 498, 32,1))

model.fit(X_train, y_train, epochs=5)

```

```python
model.evaluate(X_test, y_test )
```

Predicciones

```python
y_pred= model.predict_classes(X_test)
y_pred = y_pred.reshape(y_test.shape[0])

```

```python
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels

def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    fig, ax = plt.subplots(figsize=(10, 10))

    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.5f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

```python
unique_y = np.unique(y_train) 
plot_confusion_matrix(y_test.astype('int32'),y_pred.astype('int32'),classes=unique_y.astype('int32'))
```
